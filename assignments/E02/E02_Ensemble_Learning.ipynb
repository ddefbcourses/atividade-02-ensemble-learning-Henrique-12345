````python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

def load_data(seed=42):
    X, y = fetch_openml(
        "Fashion-MNIST",
        version=1,
        as_frame=False,
        return_X_y=True
    )
    
    y = y.astype(int)
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,
        random_state=seed,
        stratify=y
    )
    
    return X_train, X_test, y_train, y_test
````

Não é necessário normalizar os dados ao utilizar modelos baseados em árvores de decisão, como o DecisionTreeClassifier. Isso ocorre porque esses modelos realizam divisões baseadas em comparações de valores das features, e não dependem de métricas de distância ou escala dos dados.

No entanto, para modelos baseados em distância ou otimização, como KNN, SVM ou redes neurais, a normalização é importante para garantir melhor desempenho e estabilidade no treinamento.

# Questão 4

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados
- Treinar o modelo escolhido (`rf` ou `ab`)
- Avaliar o modelo
- Retornar a acurácia

**Solução**:

````python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


def load_data(seed=42):
    X, y = fetch_openml(
        "Fashion-MNIST",
        version=1,
        as_frame=False,
        return_X_y=True
    )
    
    y = y.astype(int)
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,
        random_state=seed,
        stratify=y
    )
    
    return X_train, X_test, y_train, y_test


def train_random_forest(X_train, y_train, seed=42):
    model = RandomForestClassifier(
        n_estimators=100,
        random_state=seed,
        n_jobs=-1
    )
    
    model.fit(X_train, y_train)
    return model


def train_adaboost(X_train, y_train, seed=42):
    base_estimator = DecisionTreeClassifier(
        max_depth=1,
        random_state=seed
    )
    
    model = AdaBoostClassifier(
        estimator=base_estimator,
        n_estimators=50,
        random_state=seed
    )
    
    model.fit(X_train, y_train)
    return model


def evaluate(model, X_test, y_test):
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    
    print("Acurácia:", acc)
    return acc


def run_pipeline(model_type="rf", seed=42):
    X_train, X_test, y_train, y_test = load_data(seed)
    
    if model_type == "rf":
        model = train_random_forest(X_train, y_train, seed)
    elif model_type == "ab":
        model = train_adaboost(X_train, y_train, seed)
    else:
        raise ValueError("model_type deve ser 'rf' ou 'ab'")
    
    acc = evaluate(model, X_test, y_test)
    return acc


acc_rf = run_pipeline("rf", 42)
acc_ab = run_pipeline("ab", 42)

print("Random Forest:", acc_rf)
print("AdaBoost:", acc_ab)
````

**Em qual profundidade começa o overfitting?**
**Por que a árvore consegue 100% no treino quando max_depth=None?**

O overfitting começa aproximadamente quando a profundidade da árvore ultrapassa 3 ou 4, pois a acurácia no conjunto de treino continua aumentando enquanto a acurácia no conjunto de teste deixa de melhorar ou começa a diminuir.

Quando max_depth=None, a árvore não possui restrições de crescimento e continua realizando divisões até separar completamente os dados de treino. Isso permite que o modelo atinja 100% de acurácia no treino, pois ele memoriza os dados, mas prejudica sua capacidade de generalização para novos dados, caracterizando overfitting.

# Questão 5

Execute o pipeline para ambos os modelos:

- Random Forest
- AdaBoost

## Apresente:
- Acurácia, Precisão, Recall e F1-Score de cada modelo

## Responda:
- Qual modelo apresentou melhor desempenho inicial?

**Solução**:

````python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


def load_data(seed=42):
    X, y = fetch_openml(
        "Fashion-MNIST",
        version=1,
        as_frame=False,
        return_X_y=True
    )
    
    y = y.astype(int)
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,
        random_state=seed,
        stratify=y
    )
    
    return X_train, X_test, y_train, y_test


def train_random_forest(X_train, y_train, seed=42):
    model = RandomForestClassifier(
        n_estimators=100,
        random_state=seed,
        n_jobs=-1
    )
    model.fit(X_train, y_train)
    return model


def train_adaboost(X_train, y_train, seed=42):
    base = DecisionTreeClassifier(max_depth=1, random_state=seed)
    
    model = AdaBoostClassifier(
        estimator=base,
        n_estimators=50,
        random_state=seed
    )
    model.fit(X_train, y_train)
    return model


def evaluate(model, X_test, y_test):
    y_pred = model.predict(X_test)
    
    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average="weighted")
    recall = recall_score(y_test, y_pred, average="weighted")
    f1 = f1_score(y_test, y_pred, average="weighted")
    
    print("Acurácia:", acc)
    print("Precisão:", precision)
    print("Recall:", recall)
    print("F1-score:", f1)
    
    return acc, precision, recall, f1


def run_pipeline(model_type="rf", seed=42):
    X_train, X_test, y_train, y_test = load_data(seed)
    
    if model_type == "rf":
        model = train_random_forest(X_train, y_train, seed)
    elif model_type == "ab":
        model = train_adaboost(X_train, y_train, seed)
    else:
        raise ValueError("model_type deve ser 'rf' ou 'ab'")
    
    return evaluate(model, X_test, y_test)


print("=== Random Forest ===")
rf_metrics = run_pipeline("rf", 42)

print("\n=== AdaBoost ===")
ab_metrics = run_pipeline("ab", 42)
````

O modelo Random Forest apresentou melhor desempenho inicial, com valores superiores de acurácia, precisão, recall e F1-score em comparação ao AdaBoost. Isso ocorre porque o Random Forest utiliza múltiplas árvores profundas, sendo mais adequado para capturar padrões complexos presentes no dataset Fashion MNIST.

Por outro lado, o AdaBoost utiliza modelos fracos (árvores rasas), o que limita sua capacidade de aprendizado em dados de alta dimensionalidade, resultando em desempenho inferior.

# Questão 8

Varie pelo menos um hiperparâmetro em cada modelo:

- Random Forest: `n_estimators`
- AdaBoost: `n_estimators`

## Analise:
- O desempenho muda significativamente?

## Responda:
- Qual modelo é mais sensível a mudanças?

````python
def experiment_n_estimators():
    X_train, X_test, y_train, y_test = load_data(42)
    
    valores = [10, 50, 100, 200]
    
    print("=== Random Forest ===")
    for n in valores:
        model = RandomForestClassifier(
            n_estimators=n,
            random_state=42,
            n_jobs=-1
        )
        model.fit(X_train, y_train)
        
        acc = accuracy_score(y_test, model.predict(X_test))
        print(f"n_estimators={n} -> Acurácia: {acc}")
    
    
    print("\n=== AdaBoost ===")
    for n in valores:
        base = DecisionTreeClassifier(max_depth=1, random_state=42)
        
        model = AdaBoostClassifier(
            estimator=base,
            n_estimators=n,
            random_state=42
        )
        model.fit(X_train, y_train)
        
        acc = accuracy_score(y_test, model.predict(X_test))
        print(f"n_estimators={n} -> Acurácia: {acc}")


experiment_n_estimators()
````

Ao variar o hiperparâmetro n_estimators, observa-se que o desempenho do Random Forest melhora levemente com o aumento do número de árvores, mas tende a estabilizar após certo ponto, não apresentando mudanças significativas.

Já o AdaBoost apresenta maior variação no desempenho, podendo melhorar ou até piorar conforme o número de estimadores aumenta.

Dessa forma, o AdaBoost é mais sensível a mudanças nesse hiperparâmetro, enquanto o Random Forest é mais robusto e estável.
